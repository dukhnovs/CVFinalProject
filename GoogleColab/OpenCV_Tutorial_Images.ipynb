Using Matplotlib

```
from matplotlib import pyplot as plt

image_path = "Images/17_00000001.jpg"
img = cv2.imread(image_path, 1) # 0 = grayscale, 1 = color, -1 = as is including alpha channel, img.shape = (240, 427, 3)
img2 = img[:,:,::-1] # matplot lib uses RGB but opencv uses BGR so we need to flip the red and blue channels

plt.figure(figsize=(10,10))
plt.subplot(211)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
# plt.show()

plt.subplot(212)
plt.imshow(img2, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
