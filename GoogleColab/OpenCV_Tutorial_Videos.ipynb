Read a Video

```
path = "Videos/Study clip 017.mpg"

cap = cv2.VideoCapture(path) # pass a 0 to capture from camera 0 or a file path to an existing video


while(cap.isOpened()):
    # Capture frame-by-frame
    ret, frame = cap.read() # ret is a bool (True/False) if frame is read correctly, frame.shape is (720, 1280, 3)
    # print("ret", ret, "frame", frame.shape)
    if ret == False:
      break
    frame[:,:,:] = frame[:,:,::-1] # matplot lib uses RGB but opencv uses BGR so we need to flip the red and blue channels

    plt.subplot(111)
    im = plt.imshow(frame, cmap = 'gray', interpolation = 'bicubic')
    plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
    plt.show()



# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

```
import numpy as np
import cv2

cap = cv2.VideoCapture(path)

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('Videos/output.avi',fourcc, 20.0, (640,480))

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        frame = cv2.flip(frame,0)

        # write the flipped frame
        out.write(frame)
    else:
        break

# Release everything if job is finished
cap.release()
out.release()
cv2.destroyAllWindows()
```
